```python
import os
import sys
import shutil

!git clone https://github.com/lucminhduc3108/Retrieval-ABSA.git /kaggle/working/repo
os.chdir('/kaggle/working/repo')
sys.path.insert(0, '/kaggle/working/repo')
print('Working dir:', os.getcwd())
```

```python
NB1_INPUT = '/kaggle/input/p3s2-embedding'
assert os.path.exists(NB1_INPUT), f'Dataset not found: {NB1_INPUT}'

# Copy embedding checkpoint
os.makedirs('checkpoints/embedding', exist_ok=True)
shutil.copy(f'{NB1_INPUT}/embedding_best.pt', 'checkpoints/embedding/best.pt')
print(f'Embedding ckpt: {os.path.getsize("checkpoints/embedding/best.pt") / 1e6:.1f} MB')

# Copy processed data
os.makedirs('data/processed', exist_ok=True)
for f in os.listdir(f'{NB1_INPUT}/processed'):
    shutil.copy(f'{NB1_INPUT}/processed/{f}', f'data/processed/{f}')
    print(f'data/processed/{f}')
```

## 1. Build Index

```python
!python scripts/03_build_index.py \
    --embedding_ckpt checkpoints/embedding/best.pt \
    --input data/processed/classification.jsonl \
    --bio_input data/processed/bio_tagging.jsonl \
    --out_dir indexes/
```

```python
for f in os.listdir('indexes/'):
    size = os.path.getsize(f'indexes/{f}') / 1e6
    print(f'indexes/{f}: {size:.1f} MB')
```

## 2. Save Outputs

```python
output_dir = '/kaggle/working/outputs_p3s2_nb2'
os.makedirs(output_dir, exist_ok=True)

# Index files
shutil.copytree('indexes', os.path.join(output_dir, 'indexes'), dirs_exist_ok=True)
for f in os.listdir(os.path.join(output_dir, 'indexes')):
    print(f'indexes/{f}')

print(f'\nOutputs saved to {output_dir}')
print('Upload this as Kaggle dataset: p3s2-index')
```
